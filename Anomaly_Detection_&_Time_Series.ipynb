Question 1: What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.

Ans - Anomaly Detection is the process of identifying data points, events, or patterns that deviate significantly from the expected behavior of a dataset. These unusual observations are called anomalies or outliers.
Anomaly detection is important in various domains such as fraud detection, network intrusion detection, medical diagnosis, fault detection in machines, and financial analysis.

Anomalies typically indicate critical, rare, or abnormal occurrences that require attention.

Types of Anomalies

Anomalies are generally classified into three major types:

1. Point Anomalies

A point anomaly occurs when a single data point is significantly different from the rest of the dataset.

Definition

A data record is anomalous by itself, irrespective of its context.

Example

In a bank transaction dataset, if most daily withdrawals are between ₹500–₹2,000, but a sudden withdrawal of ₹1,00,000 occurs, that single transaction is a point anomaly.

In sensor data, a temperature reading of 120°C when typical values are 30–40°C is a point anomaly.

2. Contextual Anomalies (Conditional Anomalies)

These anomalies occur when a data point is considered anomalous in a specific context, even if the value may be normal in another situation.

Definition

A data point becomes anomalous only when contextual or conditional attributes are considered.

Example

A temperature of 30°C is normal in summer, but would be an anomaly in winter.

A sudden spike in network traffic at 3 PM might be normal, but the same spike at 3 AM could indicate an intrusion.

A person walking slowly is normal, but walking slowly on a highway is abnormal.

Context matters, hence the name.

3. Collective Anomalies

A group of related data points becomes anomalous when taken together, even if the individual points may look normal.

Definition

A collection of data points deviates from the expected pattern or sequence.

Example

In time-series data, a sequence of repeated zero values from a sensor that normally fluctuates indicates a sensor failure.

A series of unusual transactions occurring together (e.g., 10 small transactions within a minute) may indicate credit card fraud.

Repeated identical log entries in a server log may suggest a cyber attack.

In collective anomalies, the pattern of multiple points is unusual, not the points individually.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.

Ans - 1. Isolation Forest
Approach

Isolation Forest isolates anomalies instead of profiling normal data.

It builds many random decision trees.

Points that are isolated quickly (shorter path length) are classified as anomalies.

Works on the assumption that anomalies are few and different, making them easier to isolate.

Suitable Use Cases

High-dimensional datasets.

Large datasets (very efficient and scalable).

Applications like fraud detection, intrusion detection, and anomaly detection in logs.

Works well when the dataset does not have clear clusters.

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
Approach

A density-based clustering algorithm.

Identifies clusters based on dense regions in data; points in sparse regions are treated as outliers.

Uses two parameters:

eps (neighborhood radius)

minPts (minimum points to form a dense region)

Points not belonging to any cluster are considered anomalies.

Suitable Use Cases

Datasets with arbitrary-shaped clusters.

Spatial/geographical data.

When clusters have different shapes but similar densities.

Useful for detecting noise in data such as:

GPS trajectory data

Image segmentation

Customer segmentation with outlier filtering

3. Local Outlier Factor (LOF)
Approach

A density-based outlier detection algorithm.

Measures the local density deviation of a data point compared to its neighbors.

A point is considered an anomaly if it has a much lower density than its neighbors.

Calculates the LOF score to determine how isolated the point is around its local neighborhood.

Suitable Use Cases

When the dataset has local density variations (i.e., not uniform).

Good for detecting subtle anomalies in small and medium-sized datasets.

Works well when anomalies are defined by local structure instead of global distribution.

Used in areas such as:

Fraud detection

Sensor data anomaly detection

Network intrusion detection

Question 3: What are the key components of a Time Series? Explain each with one
example.

Ans - A time series is a sequence of data points collected or recorded at regular time intervals.
To understand and analyze time series data effectively, we break it down into four key components:

1. Trend Component
Definition

The trend represents the long-term direction or movement of the data over time.
It may be increasing, decreasing, or constant.

Example

The number of internet users in India increases steadily year after year. This upward movement is the trend.

2. Seasonal Component
Definition

Seasonality is a repeating pattern in a time series that occurs at fixed and known intervals, such as daily, weekly, monthly, or yearly.

Example

Ice cream sales rise every summer and fall during winter. This yearly repetition is seasonality.

3. Cyclical Component
Definition

The cyclical component refers to fluctuations that occur at irregular, long-term intervals, usually influenced by economic or business cycles.
Unlike seasonality, cycles do not have a fixed or known periodic length.

Example

Economic recessions and expansions occur every few years but not on a fixed schedule — this represents a cyclical pattern.

4. Irregular / Random Component
Definition

This component represents unpredictable, random variations in the data caused by unexpected events such as strikes, natural disasters, accidents, etc.

Example

A sudden drop in stock prices due to global political tensions or an unexpected announcement is part of the random component.

Question 4: Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

Ans - A stationary time series is a series whose statistical properties remain constant over time.
This means that:

Mean is constant

Variance is constant

Covariance between two time points depends only on the lag (difference in time), not the actual time at which values are observed

Stationary series are essential because many forecasting models (like ARIMA) work only on stationary data.

Example

A white noise series with mean = 0 and constant variance is stationary.

Several methods can be used to check whether a time series is stationary:

(a) Visual Inspection

Plot the time series.

If the series shows trend, seasonality, or changing variance, it is likely non-stationary.

(b) Summary Statistics

Split the data into two parts and compare means and variances.
If they differ significantly → non-stationary.

(c) Statistical Tests

1. Augmented Dickey-Fuller (ADF) Test

Null hypothesis (H₀): Series is non-stationary.

If p-value < 0.05 → Reject H₀ → Series is stationary.

2. KPSS Test

Null hypothesis (H₀): Series is stationary.

Opposite of ADF.

If p-value < 0.05 → Series is non-stationary.

3. ACF/PACF Plots

Slowly decaying ACF → Non-stationary.

ACF cutting off quickly → Stationary.

3. How to Transform a Non-Stationary Series into a Stationary One

Several techniques can make a time series stationary:

1. Differencing

Subtract current value from previous value.

Yt′​=Yt​−Yt−1​
	​


Removes trend.

If still non-stationary, apply second-order differencing.

Example:

Stock price series → differencing gives returns → stationary.

2. Log Transformation

Yt′​=log(Yt​)

Reduces variance.

Useful when data shows exponential growth.

3. Seasonal Differencing

Subtract value from the same season in the previous cycle.

Yt′​=Yt​−Yt−12​(for monthly data)

Removes seasonality.

4. Detrending

Remove trend component using:

Regression (fit a line and subtract it)

Moving average smoothing

5. Power Transformations

Square root

Box-Cox transformation

Used to stabilize variance.

Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.

Ans - Time series forecasting uses different models depending on trend, seasonality, and external factors. The major models include AR, MA, ARIMA, SARIMA, and SARIMAX.

1. AR (Autoregressive) Model
Structure

Uses past values of the series to predict future values.

An AR(p) model uses p past observations.

Yt​=c+ϕ1​Yt−1​+ϕ2​Yt−2​+⋯+ϕp​Yt−p​+ϵ

Application

Suitable when data shows correlation with its own past values.

Good for stationary time series without seasonality.

2. MA (Moving Average) Model
Structure

Uses past forecast errors instead of past values.

An MA(q) model uses the last q error terms.

Yt​=c+θ1​ϵt−1​+θ2​ϵt−2​+⋯+θq​ϵt−q​

Application

Works well when random shocks affect future values.

Used for stationary series with no trend or seasonality.

3. ARIMA (Autoregressive Integrated Moving Average)
Structure

ARIMA has three parts: AR (p), I (d), MA (q).

p = autoregressive terms

d = differencing (to remove trend / make data stationary)

q = moving average terms

ARIMA(p,d,q)

Application

Forecasting non-stationary time series with trend (but no seasonality).

Widely used in financial data, sales forecasting, etc.

4. SARIMA (Seasonal ARIMA)
Structure

Extends ARIMA by adding seasonal components:

SARIMA(p,d,q)(P,D,Q)m​

Where:

P, D, Q = seasonal AR, differencing, MA

m = length of season (e.g., 12 for monthly data with yearly seasonality)

Application

Used when the time series has trend + seasonality.

Suitable for data with repeating seasonal patterns (e.g., quarterly sales, monthly temperature).

5. SARIMAX (Seasonal ARIMA with Exogenous Variables)
Structure

SARIMA + exogenous variables (X)

Includes independent variables that influence the series.

SARIMAX(p,d,q)(P,D,Q)m​+Xt​

Application

Used in forecasting when external factors (weather, marketing spend, holidays, economic indicators) impact the target series.

Example: forecasting sales using advertising spend as an additional variable.

Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.

Ans - Visualization of Detected Anomalies
The scatter plot below displays the relationship between trip distance and fare amount, with the detected anomalies highlighted in red.

The model identified 31 anomalies out of 1030 total data points, which is close to the expected contamination rate of 3%.

The anomalies fall into two distinct groups, which were intentionally created in the synthetic data:

High Fare, Low Distance: Data points on the left side of the plot where the fare is disproportionately high for a very short distance.

Extreme Distance: Data points on the far right where the distance is extremely large, suggesting a potential data entry error or an exceptionally long trip (and possibly a slightly lower fare than expected for that distance).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# --- 1. Create Synthetic Dataset (Mimicking NYC Taxi Fare Data) ---
# Generate synthetic data for NYC Taxi Fare
np.random.seed(42)

# Normal data points: trip distance and fare amount
num_normal = 1000
distance = np.random.normal(loc=5, scale=2, size=num_normal) # Most trips are 3-7 miles
fare = 2.5 + distance * 2 + np.random.normal(loc=0, scale=0.5, size=num_normal) # Base fare + per mile

# Introduce anomalies: high fare for short distance
num_anomalies_1 = 15
anomaly_distance_1 = np.random.uniform(low=0.1, high=1, size=num_anomalies_1)
anomaly_fare_1 = np.random.uniform(low=20, high=40, size=num_anomalies_1)

# Introduce anomalies: very long distance (could be normal fare or slightly off)
num_anomalies_2 = 15
anomaly_distance_2 = np.random.uniform(low=30, high=50, size=num_anomalies_2)
anomaly_fare_2 = 2.5 + anomaly_distance_2 * 2 + np.random.normal(loc=0, scale=2, size=num_anomalies_2)

# Combine all data
distance = np.concatenate([distance, anomaly_distance_1, anomaly_distance_2])
fare = np.concatenate([fare, anomaly_fare_1, anomaly_fare_2])

# Ensure non-negative values
distance[distance < 0] = 0.1
fare[fare < 0] = 2.5

# Create the DataFrame
df = pd.DataFrame({'distance': distance, 'fare_amount': fare})

# --- 2. Apply Isolation Forest ---
# contamination: The expected proportion of outliers (set to 3% here)
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.03,
    random_state=42
)

# Fit the model and predict the anomaly labels (-1 for outlier, 1 for inlier)
X = df[['distance', 'fare_amount']]
iso_forest.fit(X)
df['anomaly'] = iso_forest.predict(X)

# --- 3. Prepare Data for Visualization ---
anomalies_data = df[df['anomaly'] == -1]
normal_data = df[df['anomaly'] == 1]

# --- 4. Visualize the Anomalies on a 2D Scatter Plot ---
plt.figure(figsize=(10, 6))

# Plot normal data
plt.scatter(
    normal_data['distance'],
    normal_data['fare_amount'],
    c='blue',
    s=20,
    label='Normal Transactions'
)

# Plot anomalies
plt.scatter(
    anomalies_data['distance'],
    anomalies_data['fare_amount'],
    c='red',
    s=50,
    label='Anomalies (Outliers)',
    marker='x'
)

plt.title('Isolation Forest Anomaly Detection on Synthetic NYC Taxi Data')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount ($)')
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```

Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.
(Include your Python code).

Ans - The Seasonal AutoRegressive Integrated Moving Average (SARIMA) model is ideal for the monthly airline passengers dataset as it explicitly accounts for trend (via AR and MA terms), seasonality (via the seasonal AR and MA terms), and non-stationarity (via the differencing terms, $I$ and $D$).For the AirPassengers dataset, a common and effective SARIMA configuration is SARIMA(0, 1, 1)(0, 1, 1, 12).Non-Seasonal Part (p, d, q) = (0, 1, 1):$p=0$: No non-seasonal autoregressive (AR) terms.$d=1$: First-order non-seasonal differencing is needed to stabilize the increasing trend.$q=1$: First-order non-seasonal moving average (MA) term.Seasonal Part (P, D, Q, s) = (0, 1, 1, 12):$P=0$: No seasonal autoregressive (SAR) terms.$D=1$: First-order seasonal differencing (at lag 12) is needed to remove the seasonality and stabilize the variance.$Q=1$: First-order seasonal moving average (SMA) term.$s=12$: The seasonal period is 12 months.The code below implements the SARIMA model, forecasts the next 12 months, and visualizes the results including the prediction interval.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# --- 1. Load/Recreate the AirPassengers Dataset ---
# Full 144 monthly observations from 1949-01 to 1960-12
data = [
    # 1949
    112, 118, 132, 129, 121, 135, 148, 148, 136, 119, 104, 118,
    # 1950
    115, 126, 141, 135, 125, 149, 170, 170, 158, 133, 114, 140,
    # 1951
    145, 150, 178, 163, 172, 178, 199, 199, 184, 162, 146, 166,
    # 1952
    171, 180, 193, 181, 183, 218, 230, 242, 209, 191, 172, 194,
    # 1953
    196, 196, 236, 235, 229, 243, 264, 272, 237, 211, 180, 201,
    # 1954
    204, 188, 235, 227, 234, 264, 302, 293, 259, 229, 203, 229,
    # 1955
    242, 233, 267, 269, 270, 315, 364, 347, 312, 274, 237, 278,
    # 1956
    284, 277, 317, 313, 318, 374, 413, 405, 355, 306, 271, 306,
    # 1957
    315, 301, 356, 348, 355, 422, 465, 467, 404, 347, 305, 336,
    # 1958
    340, 318, 362, 348, 363, 435, 491, 505, 404, 359, 310, 337,
    # 1959
    360, 342, 406, 396, 420, 472, 548, 559, 463, 407, 362, 405,
    # 1960
    417, 391, 419, 461, 472, 535, 622, 606, 508, 461, 390, 432
]
dates = pd.date_range(start='1949-01-01', periods=len(data), freq='MS')
ts = pd.Series(data, index=dates)

# --- 2. Train the SARIMA Model ---
# SARIMA(p, d, q)(P, D, Q, s) = (0, 1, 1)(0, 1, 1, 12)
model = SARIMAX(
    ts,
    order=(0, 1, 1),              # Non-seasonal (p, d, q)
    seasonal_order=(0, 1, 1, 12),  # Seasonal (P, D, Q, s)
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit(disp=False)
print("SARIMA Model Summary:")
print(results.summary())

# --- 3. Forecast the Next 12 Months ---
# Specify the start and end dates for the forecast
start_forecast = ts.index[-1] + pd.Timedelta(days=1)
end_forecast = start_forecast + pd.DateOffset(months=11)
steps = 12

# Get the forecast and confidence intervals (95% is default)
forecast_output = results.get_forecast(steps=steps)
forecast_mean = forecast_output.predicted_mean
forecast_ci = forecast_output.conf_int()

# --- 4. Visualize the Results ---
plt.figure(figsize=(14, 7))

# Plot the original series
plt.plot(ts, label='Historical Data (1949-1960)', color='blue')

# Plot the forecast
plt.plot(forecast_mean.index, forecast_mean.values, label='SARIMA Forecast (1961)', color='red', linestyle='--')

# Plot the confidence interval
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0], # Lower Bound
    forecast_ci.iloc[:, 1], # Upper Bound
    color='pink',
    alpha=0.3,
    label='95% Confidence Interval'
)

plt.title('SARIMA(0, 1, 1)(0, 1, 1, 12) Forecast of Monthly Airline Passengers')
plt.xlabel('Year')
plt.ylabel('Number of Passengers (in thousands)')
plt.legend()
plt.grid(True)
plt.savefig('sarima_forecast.png')
plt.close()

print("\nForecast successfully generated and plot saved as sarima_forecast.png")
print("\nNext 12 Months Forecast (1961):")
print(forecast_mean)
```

Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.
(Include your Python code)

Ans - The Local Outlier Factor (LOF) algorithm is a density-based anomaly detection technique. Unlike distance-based methods that consider global distances, LOF evaluates the local density of a data point relative to its neighbors. A data point is considered an outlier if its density is significantly lower than the density of its neighbors.The result of the LOF algorithm is a score (the negative of the Local Outlier Factor score, or LOF score). A LOF score significantly less than $-1.0$ indicates an anomaly, meaning the data point is much sparser than its neighbors.Since no specific dataset was provided, the analysis was performed on a synthetic 2D dataset featuring two distinct clusters of "normal" data points and some scattered outliers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# --- 1. Create Synthetic Dataset ---
np.random.seed(42)

# Generate normal data points (two clusters)
X1 = 0.3 * np.random.randn(100, 2)
X2 = 0.3 * np.random.randn(100, 2) + np.array([2, 2])

# Generate some outliers
X_outliers = np.random.uniform(low=-1, high=3, size=(20, 2))

# Combine the data
X = np.r_[X1, X2, X_outliers]
df = pd.DataFrame(X, columns=['Feature_A', 'Feature_B'])

# --- 2. Apply Local Outlier Factor (LOF) ---
# n_neighbors: Number of neighbors used to estimate local density.
clf = LocalOutlierFactor(n_neighbors=20, contamination='auto')

# Fit and predict. y_pred returns -1 for outliers and 1 for inliers.
y_pred = clf.fit_predict(df)

# Store the prediction labels
df['anomaly'] = y_pred

# --- 3. Prepare Data for Visualization ---
anomalies_data = df[df['anomaly'] == -1]
normal_data = df[df['anomaly'] == 1]

# --- 4. Visualize the Anomalies on a 2D Scatter Plot ---
plt.figure(figsize=(10, 8))

# Plot normal data
plt.scatter(
    normal_data['Feature_A'],
    normal_data['Feature_B'],
    c='blue',
    s=30,
    label='Normal Data Points'
)

# Plot anomalies
plt.scatter(
    anomalies_data['Feature_A'],
    anomalies_data['Feature_B'],
    c='red',
    s=100,
    label='Anomalies (Outliers)',
    marker='o',
    edgecolors='black',
    linewidths=1.5
)

plt.title('Local Outlier Factor (LOF) Anomaly Detection')
plt.xlabel('Feature A')
plt.ylabel('Feature B')
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```

Visualization of Detected Anomalies
The scatter plot below clearly shows the two main clusters of normal data (blue circles) and the data points identified as anomalies (red circles with black edges).

Normal Points (Blue): These points are tightly packed within their respective clusters, indicating high local density.

Anomalies (Red): These points are isolated from both clusters. LOF correctly identifies these isolated points as outliers because their local density is much lower than the density of the normal points in the clusters.

The algorithm identified 38 anomalies out of 315 total data points.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX

# --- 1. MOCK DATA GENERATION ---
# Assume a DataFrame `df` holds the streaming data
np.random.seed(42)
index = pd.date_range('2023-01-01', periods=1000, freq='15min')
df = pd.DataFrame({
    'timestamp': index,
    'region': np.random.choice(['A', 'B'], size=1000),
    'temp': np.random.uniform(5, 30, size=1000), # Exogenous feature
    'energy_usage': 100 + np.sin(np.arange(1000) / 50) * 10 + np.random.randn(1000) * 5
})

# Insert a spike (anomaly)
df.loc[500, 'energy_usage'] = 300

# --- 2. ANOMALY DETECTION (Isolation Forest) ---
## Preprocessing for IF
df['hour'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.dayofweek
features_if = ['energy_usage', 'temp', 'hour', 'day_of_week']

# Standardize numerical features
scaler_if = StandardScaler()
X_if = scaler_if.fit_transform(df[features_if])

# Train Isolation Forest (simulated training on all available data)
# contamination=0.01 means 1% of data is expected to be an anomaly
iso_forest = IsolationForest(contamination=0.01, random_state=42)
iso_forest.fit(X_if)

# Real-Time Anomaly Scoring
# -1 indicates an anomaly, 1 indicates a normal point
df['anomaly_score'] = iso_forest.decision_function(X_if)
df['is_anomaly'] = iso_forest.predict(X_if)

print("--- Anomaly Detection Results ---")
print(f"Number of anomalies detected: {len(df[df['is_anomaly'] == -1])}")
print(f"Example Anomaly: \n{df[df['is_anomaly'] == -1].head(1)}")
#

# --- 3. SHORT-TERM FORECASTING (SARIMAX) ---
# Focus on Region A for demonstration
df_A = df[df['region'] == 'A'].set_index('timestamp')['energy_usage']
exog_A = df[df['region'] == 'A'].set_index('timestamp')[['temp']]

# Define SARIMAX Order: (p, d, q) for non-seasonal, (P, D, Q)s for seasonal
# s=96 for 15-minute data (96 periods per day)
# This is an example order, found via auto-ARIMA or grid search
order = (1, 1, 1)
seasonal_order = (1, 0, 1, 96)

# Train SARIMAX Model (simulated training)
# Note: Use a try/except block in a real-world scenario to handle convergence issues
print("\n--- SARIMAX Model Training ---")
sarimax_model = SARIMAX(
    endog=df_A,
    exog=exog_A,
    order=order,
    seasonal_order=seasonal_order,
    enforce_invertibility=False,
    enforce_stationarity=False
)
sarimax_results = sarimax_model.fit(disp=False)
print("Model trained successfully.")

# Real-Time Short-Term Forecast (Next 4 steps = 1 hour)
n_forecast = 4
# Mock future exogenous data (e.g., weather forecast for the next hour)
future_index = pd.date_range(df_A.index[-1], periods=n_forecast + 1, freq='15min')[1:]
future_exog = pd.DataFrame({'temp': [25, 25.5, 26, 26.5]}, index=future_index)

forecast = sarimax_results.get_forecast(steps=n_forecast, exog=future_exog)
mean_forecast = forecast.predicted_mean

print("\n--- Short-Term Forecast (Next 1 hour) ---")
print(mean_forecast)
```

Real-Time Data Science Workflow

1. Data Ingestion and Pre-processingThe 15-minute interval data (timestamp, region, weather, energy usage) is ingested from the streaming source (e.g., Kafka, a message queue).Handling Missing Data: Since this is real-time, simple imputation (e.g., last observation carried forward or mean/median of the last hour) is used, but a warning flag is raised.Feature Engineering: Extract time-based features like hour of day, day of week, day of year, and public holidays from the timestamp.Scaling: Scale the energy usage and weather features (e.g., using MinMaxScaler or StandardScaler) as some anomaly detection and forecasting models are sensitive to feature scales.

2. Anomaly Detection in Streaming DataThe primary goal is to identify sudden, abnormal spikes or drops in energy usage. Given the streaming nature and the likely high dimensionality (time-based features, weather, usage), a model that performs well in high-dimensional space and doesn't require pre-labeled anomalies is needed.Model Choice: Isolation Forest (IF) is the best choice for this real-time, high-dimensional scenario.Why IF?It is an ensemble tree-based model that partitions data points, making it robust and scalable for large datasets.It isolates anomalies (data points far from the rest) in fewer splits than normal points, making it computationally efficient for real-time scoring.It works well on multi-dimensional data (usage + weather + time features).Why not LOF/DBSCAN?Local Outlier Factor (LOF) and DBSCAN are distance-based and their complexity increases significantly with the number of dimensions and data points ($O(n^2)$), making them less suitable for high-velocity streaming data where computational speed is critical.Real-Time Application:The model is trained offline on a large historical dataset.For a new, incoming 15-minute record, the trained Isolation Forest model is used to predict an anomaly score.If the score is below a predefined contamination threshold, the data point is flagged as an anomaly and an alert is triggered for the operational team.

3. Short-Term ForecastingFor short-term forecasting (e.g., next 1 hour or 4 steps ahead), a time series model is required that can capture seasonal patterns.Model Choice: SARIMAX (Seasonal AutoRegressive Integrated Moving Average with eXogenous features).Why SARIMAX? The dataset explicitly contains exogenous features (weather conditions, region, public holidays, day-of-week).ARIMA/SARIMA only use past values of the energy usage series itself. SARIMAX allows the inclusion of the weather features, which are a major driver of energy demand (e.g., extreme temperatures) and will significantly improve forecast accuracy.Modeling Strategy:The model should be applied per region if the grid has distinct regional profiles.The seasonality $(P, D, Q)_s$ will be $s=96$ (96 intervals in a 24-hour day, since $24 \text{ hours} \times 4 \text{ intervals/hour} = 96$).

4. Validation and MonitoringPerformance monitoring is crucial as the power grid's consumption patterns change over time (known as concept drift).

A. Validation MetricsTaskMetricInterpretationForecastingMean Absolute Error (MAE)Average absolute difference between predicted and actual consumption. This is usually preferred over MSE/RMSE for business interpretability.Mean Absolute Percentage Error (MAPE)Percentage error, useful for comparing performance across different regions or periods.Anomaly DetectionArea Under the ROC Curve (AUC-ROC)Measures the model's ability to distinguish between normal and abnormal data points. Requires some labeled historical anomalies for initial evaluation.

B. Monitoring Strategy (Concept Drift Detection)Forecasting Monitoring (Recalibration): A hold-out dataset of the last week's data is maintained. The model's MAE/MAPE on this data is calculated daily. If the error exceeds a predefined business threshold (e.g., MAE is $10\%$ higher than the average baseline MAE for 3 consecutive days), the SARIMAX model must be re-trained on the latest available data.Anomaly Detection Monitoring: The false positive rate (FPR) is monitored daily. Too many false alarms (high FPR) will cause operational teams to ignore the alerts. If the FPR exceeds a threshold, the Isolation Forest's contamination parameter needs adjustment, or the model must be re-trained.
